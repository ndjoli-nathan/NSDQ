```python
from MyPackages import NSDQ
from MyPackages import  BIS_CBPR #Central Bank Policy Rates from BIS
```

```python
Apple = NSDQ.Asset('AAPL','stocks') #Create an asset object
```

```python
Apple.Informations() #Print informations about the asset
```

```python
Apple.HistoricalData() #Get the historical data of the asset
```

```python
Apple.RealTime() #Get the real time data of the asset
```

```python
Apple.Options("raw", 'oprac') #Get raw options data of the asset
```

```python
r = BIS_CBPR.CentralBankPolicyRates('United States') #Get the Central Bank Policy Rates of the United States from BIS
```

```python
Apple.Options("processed", 'oprac', 'callput', r) #Get processed options data of the asset (Including Implied Volatility & Greeks calculated using Black-Scholes Merton model)
```

```python
IVS = Apple.ImpliedVolatilitySurface('callput', r) #Create an Implied Volatility Surface object for the asset
```

```python
IVS.Plot(SmoothingFactor=1, Granularity=256, ShowScatter=True) #Plot the Implied Volatility Surface
```
